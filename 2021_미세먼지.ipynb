## **중국 & 서울**

```python
Seoul_2018 = Seoul[Seoul['year']==2018]
Seoul_2019 = Seoul[Seoul['year']==2019]
Seoul_2020 = Seoul[Seoul['year']==2020]
```

```python
china_2018_month = china_2018.groupby('month')['mean'].agg(**{'pm10':'mean'})
china_2019_month = china_2019.groupby('month')['mean'].agg(**{'pm10':'mean'})
china_2020_month = china_2020.groupby('month')['mean'].agg(**{'pm10':'mean'})
```

```python
fig, ax = plt.subplots(nrows=3, figsize=(10,15))
plt.subplots_adjust(hspace = 0.3)
sns.set_style("whitegrid")

axes = sns.lineplot(x="month", y="pm10", data= Seoul_2018, color="red", ax=ax[0])
axes = sns.lineplot(x=china_2018_month.index, y="pm10", data= china_2018_month, color="Blue", ax=ax[0])
axes1 = sns.lineplot(x="month", y="pm10", data= Seoul_2019, color="red", ax=ax[1])
axes1 = sns.lineplot(x=china_2019_month.index, y="pm10", data= china_2019_month, color="Blue", ax=ax[1])
axes2 = sns.lineplot(x="month", y="pm10", data= Seoul_2020, color="red", ax=ax[2])
axes2 = sns.lineplot(x=china_2020_month.index, y="pm10", data= china_2020_month, color="Blue", ax=ax[2])
axes.legend(labels = ['Seoul','Shandong'],loc = 'upper center', fontsize=15)
axes1.legend(labels = ['Seoul','Shandong'],loc = 'upper center', fontsize=15)
axes2.legend(labels = ['Seoul','Shandong'],loc = 'upper center', fontsize=15)
axes.set_title("2018 Seoul & Shandong", fontsize=15)
axes1.set_title("2019 Seoul & Shandong", fontsize=15)
axes2.set_title("2020 Seoul & Shandong", fontsize=15)
```
