# K - 평균
- 각각의 픽셀값(3차원 -> 1차원 배열)의 평균 구함.
  + 픽셀의 평균값을 활용하여 사과, 바나나, 파인애플에 근사한 이미지를 추출
-평균값 구하는 방법
  + K - 평균 알고리즘(K-Means) 알고리즘
  + 평균값 = Cluster Center = Centroid

- 3차원 배열을 2차원 배열로 만들어 보자
  + 3차원 (샘플개수, 너비, 높이)
  + ** 2차원 (샘플개서, 너비 * 높이) **

```python
fruits_2d = fruits.reshape(-1, 100 * 100) 
fruits_2d.shape
```

- k - 평균 알고리즘 활용
  + 비지도 학습에서는 타겟(종속변수)이 없다(다시한번기억!)

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3, random_state=42)
km.fit(fruits_2d)
```

- 모형 학습 후 labels_

```python
print(km.labels_)
```

- 직접 샘플의 개수 확인

```python
print(np.unique(km.labels_, return_counts = True))
```

- 그래프를 그려보자
  + 이미지가 섞여 있는데 분류해서 보여줌

```python
import matplotlib.pyplot as plt

def draw_fruits(arr, ratio=1):
    n = len(arr)    # n은 샘플 개수입니다
    # 한 줄에 10개씩 이미지를 그립니다. 샘플 개수를 10으로 나누어 전체 행 개수를 계산합니다. 
    rows = int(np.ceil(n/10))
    # 행이 1개 이면 열 개수는 샘플 개수입니다. 그렇지 않으면 10개입니다.
    cols = n if rows < 2 else 10
    fig, axs = plt.subplots(rows, cols, 
                            figsize=(cols*ratio, rows*ratio), squeeze=False)
    for i in range(rows):
        for j in range(cols):
            if i*10 + j < n:    # n 개까지만 그립니다.
                axs[i, j].imshow(arr[i*10 + j], cmap='RdBu_r')
            axs[i, j].axis('off')
    plt.show()
```

```python
draw_fruits(fruits[km.labels_ == 0])
```

- 최적의 K - 평균 찾기

```python
inertia = []
for k in range(2, 7):
  km = KMeans(n_clusters = k, random_state=42)
  km.fit(fruits_2d) # 비지도 학습이니까 타겟없이 적고
  inertia.append(km.inertia_) # inertia에 km.inertia를 추가하고
plt.plot(range(2, 7), inertia) # 그래프 2에서 7 + inertia 해서 표시
plt.show()
```

#### PCA(주성분 분석)
--> 이미지의 경우 같은 특성이라서 바로 PCA로 갈 수 있지만, 엑셀 파일 같이 각 특성이 다른 경우는 데이터 전처리와 표준화를 시켜준 다음에 PCA 작업에 들어가줘야 한다.

- 차원 축소의 개념 : 일부 특성을 선택하여 데이터 크기를 줄임
  + 머신러닝 측면 -> 과대적합 방지 & 성능 향상
- PCA 개념
  + 특성이 많을 때, 수치형 데이터가 많을 때. 하지만 문자형에서 쓰는 건 추천 하지 않음.
- 정형데이터에서도 활용 가능
  + 문자열 데이터, 수치형 데이터(연속형 데이터, 비연속형 데이터)
- 알고리즘을 구성 할 때, 필요한 데이터 픽셀 수
  + 300 * 10000개 픽셀 --> 300 * pc 10의 주성분으로 줄임
  + 기존 1시간 걸리던게 10분 걸렸고, 분류도 잘 되었다.

데이터 불러왔음 (챕터6_1)

```python
# 이미지 불러오기 및 shape확인
import numpy as np
fruits = np.load('fruits_300.npy')
fruits_2d = fruits.reshape(-1, 100 * 100)
fruits_2d.shape
```

- sklearn.decomposition 모듈
  + n_components : 성분수. 즉, PCA를 몇 개의 성분으로 하겠다.

```python
# 압축 시키는 단계(50개로)
from sklearn.decomposition import PCA
pca = PCA(n_components = 50) # PCA 50개 성분으로 300 * 10000 픽셀값을 압축
pca.fit(fruits_2d)
```

```python
import matplotlib.pyplot as plt

def draw_fruits(arr, ratio=1):
    n = len(arr)    # n은 샘플 개수입니다
    # 한 줄에 10개씩 이미지를 그립니다. 샘플 개수를 10으로 나누어 전체 행 개수를 계산합니다. 
    rows = int(np.ceil(n/10))
    # 행이 1개 이면 열 개수는 샘플 개수입니다. 그렇지 않으면 10개입니다.
    cols = n if rows < 2 else 10
    fig, axs = plt.subplots(rows, cols, 
                            figsize=(cols*ratio, rows*ratio), squeeze=False)
    for i in range(rows):
        for j in range(cols):
            if i*10 + j < n:    # n 개까지만 그립니다.
                axs[i, j].imshow(arr[i*10 + j], cmap='gray_r')
            axs[i, j].axis('off')
    plt.show()
draw_fruits(pca.components_.reshape(-1, 100, 100))
```

```python

fruits_pca = pca.transform(fruits_2d)
print(fruits_pca.shape)
```

- 원본 데이터 재구성
  + 10000개의 데이터에서 특성을 50개로 줄임. ↑
  + 100% 재구성은 어렵지만 쓸만하게 재구성 하기. ↓
- 축소된 공간에서 데이터 포인트를 가져와서 (물론 정보 손실과 함께) 원래 공간으로 되돌리는 수식. -> fruits_inverse = pca.inverse_transform(fruits_pca)

```python
fruits_inverse = pca.inverse_transform(fruits_pca)
print(fruits_inverse.shape)

fruits_reconstruct = fruits_inverse.reshape(-1, 100, 100)
print(fruits_reconstruct.shape)

fruits_reconstruct = fruits_inverse.reshape(-1, 100, 100)
for start in [0, 100, 200]:
    draw_fruits(fruits_reconstruct[start:start+100])
    print("|n")
```

```python
plt.plot(pca.explained_variance_ratio_)
plt.show()
```

- ↑ 처음 10개의 주성분이 대부분의 분산을 표현 하고, 11개~50개까지는 잘 설명이 되지 않음.

```python
print(np.sum(pca.explained_variance_ratio_))
```

#### 다른 알고리즘과 함께 사용 - 지도 학습
